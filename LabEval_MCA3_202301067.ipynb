```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_name = "company_sales_data.csv"
data = pd.read_csv(dataset_name)
```

```python
face_sales_data = data[['month_number', 'facecream', 'facewash']]
```

```python
face_sales_data.plot(x='month_number', kind='bar', figsize=(10, 6))
plt.title('Face Cream and Facewash Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['total_profit'].plot(kind='hist', bins=5, title='Total Profit Histogram')
plt.xlabel('Total Profit Range')
plt.ylabel('Frequency')
plt.show()
```
